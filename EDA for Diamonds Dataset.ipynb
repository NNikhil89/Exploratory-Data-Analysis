##### Detailed representation of carat using describe

* From the description of carat field we can find that mean and median are closer.
* There is a wide difference between Quartile value at 75% is 1.04 and maximum value of 5.01 which may result in presence of outliers which we will confirm from boxplots

```python
# defining a matplotlib figure
plt.figure(1)
# Adjusting the figure size using figure(figsize=())
plt.figure(figsize=(20,6))
plt.subplot(121)
#Using matplotlib hist to plot the histogram of carat
plt.hist(data_frame.carat);

plt.subplot(122)
#Using .plot to plot boxplot
data_frame["carat"].plot(kind='box')
plt.show()
```

The above boxplot clearly signifies the presence of outliers in the carat attribute with upper whisker value around 2 but the maximum value in the dataset being at 5.01

```python
plt.subplot(121)
#Using seaborn histplot to plot the carat 
sns.distplot(data_frame.carat, kde = True)

plt.subplot(122)
#Using seaborn histplot to plot the carat 
sns.boxplot(data_frame.carat)
plt.show()
```

* The above seaborn boxplot also confirms the presence of outliers with upper whisker at 2 and the presence of outlier near 5.
* The histogram suggests the distribution is right-skewed with more data points with lower values, and fewer data points with higher values, causing the distribution to be stretched out towards the higher end of the scale.

### Proportion of diamonds have a clarity of Sl1, Sl2, or VS2

* From the above barplot we can interpret that the proportion of SI1 SI2 and VS2 from the given dataset is between 0.6 and 0.7 i.e.around 0.65 we can calculate the exact value as below

```python
print("Proportion SI1 SI2 VS2", proportionofSI12VS2)
```

Apart from visual appeal, we can identify that the marking of Quartiles also differs between matplotlib boxplot and seaborn boxplot. For example for the clarity type 'SI2' the 50% whisker is present near 0.7 for matplotlib but for seaborn it is above 1.
* The calculation of quartiles (Q1, Q2, Q3) regions for these two plots can vary leading to differences in the placement and size of the box and whiskers. 
* Seaborn may use different methods for identifying and plotting outliers, affecting the appearance of the box plots. 
* Matplotlib typically uses linear interpolation, where Q1 is located at the 25th percentile of the data, Q3 is at the 75th percentile, and the whiskers extend to the most extreme data points within a specified range (e.g., 1.5 times the interquartile range). 
* Seaborn offers multiple interpolation methods, including "linear," "nearest," "lower," "higher," and "midpoint." 
* These methods can affect how quartiles and whiskers are calculated and positioned in the plot.
